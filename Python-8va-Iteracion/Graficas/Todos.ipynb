```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CNR100=np.loadtxt('CNRGene100')
CNR80=np.loadtxt('CNRGene80')
CNR60=np.loadtxt('CNRGene60')
CNR40=np.loadtxt('CNRGene40')
CNR20=np.loadtxt('CNRGene20')
CNR0=np.loadtxt('CNRGene0')
```

```python
VoltajesCorre= [16,18,20,22,24,26,28,30,35,40,45]
fig = plt.figure(figsize=(15,10))
plt.plot(VoltajesCorre,CNR100, 'o-', label='CNR 100%')
plt.plot(VoltajesCorre,CNR80, 'o-', label='CNR 80%')
plt.plot(VoltajesCorre,CNR60, 'o-', label='CNR 60%')
plt.plot(VoltajesCorre,CNR40, 'o-', label='CNR 40%')
plt.plot(VoltajesCorre,CNR20, 'o-', label='CNR 20%')
plt.plot(VoltajesCorre,CNR0, 'o-', label='CNR 0%')
#plt.plot(VoltajesCorre,CNRGene,'*-', label='General')
plt.xlabel('Voltajes (kV)', fontsize= 15)
plt.ylabel('Valor de CNR', fontsize= 15)
plt.xticks(VoltajesCorre, VoltajesCorre )
#plt.yticks([0,3,4,5,6,7,10,11], [3,4,5,6,7,10,11])
plt.title('Valores de CNR para todos los fantomas', fontsize= 20)
plt.grid()
plt.legend()
plt.savefig('CNRsTODOS.png')
```
